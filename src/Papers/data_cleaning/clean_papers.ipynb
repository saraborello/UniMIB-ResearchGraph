```python
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()

BASE_DIR = os.getenv("BASE_DIR")
file_path = os.path.join(BASE_DIR, 'data/raw/papers_duplicated.csv')
```

```python
# Import dataset
df = pd.read_csv(file_path, sep=",")
df
```

When opening the dataset, it is clear that there are many duplicate observations. This can be explained by the fact that many papers are written by Bicocca staff.

```python
df.drop_duplicates(subset=['Title'], keep="first", inplace=True)
df
```

There are some papers with too many authors which should be delete

```python
# Check papers with more than 25 authors

i = 1
for index, row in df.iterrows():
    title = row["Title"]
    authors = row['Authors']
    if len(authors.split(",")) > 25:
        df = df.drop(index=index)
        print(title, i)
        i += 1
```

```python
# Export the df to csv

output_path = os.path.join(BASE_DIR, 'data/processed/papers.csv')
df.to_csv(output_path, index=False)
```
