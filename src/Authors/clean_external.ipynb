```python
import pandas as pd
import pandas as pd
import csv
import requests
import google.generativeai as genai
import os
from dotenv import load_dotenv
from Orcid_ProfessorDataCollector import ProfessorDataCollector
load_dotenv()
```

```python
authors = pd.read_csv('../../data/processed/Authors.csv')
papers = pd.read_csv('../../data/processed/papers.csv')
```

```python
# Supponiamo di avere due liste
list_a = authors['ORCID ID'].tolist()  # Prima lista
list_b = papers['Authors']
all_authors = [author.strip() for authors in list_b for author in authors.split(',')]
unique_authors = list(set(all_authors))
list_b = unique_authors # Seconda lista


# Converte le liste in insiemi per il confronto
set_a = set(list_a)
set_b = set(list_b)

# Elementi in 'a' che non sono in 'b'
only_in_a = set_a - set_b

# Elementi in 'b' che non sono in 'a'
only_in_b = set_b - set_a

# Stampa i risultati
print("Elementi presenti solo in 'a':", only_in_a)
print("Elementi presenti solo in 'b':", only_in_b)
```

```python
# Normalizza le liste rimuovendo spazi e convertendo in stringhe
list_a = [str(x).strip() for x in list_a if pd.notna(x)]
list_b = [str(x).strip() for x in list_b if pd.notna(x)]

# Converte in insiemi
set_a = set(list_a)
set_b = set(list_b)

# Riconfronta
only_in_a = set_a - set_b
only_in_b = set_b - set_a

print("Elementi presenti solo in 'a':", len(only_in_a))
print("Elementi presenti solo in 'b':", len(only_in_b))
```

```python
# Filtra il DataFrame authors mantenendo solo le righe il cui ORCID non è in only_in_a
filtered_authors = authors[~authors['ORCID ID'].isin(only_in_a)]
list_c = filtered_authors['ORCID ID'].tolist()
list_c = [str(x).strip() for x in list_c if pd.notna(x)]
set_c = set(list_c)
only_in_c = set_c - set_b
filtered_authors.to_csv('Authors_small.csv')
```
